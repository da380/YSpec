```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
font = {'size'   : 20}
plt.rc('font', **font),
plt.rcParams['figure.figsize'] = [16, 8]



d = np.loadtxt('yspec.out.50')
d1 = np.loadtxt('yspec.out.ng.50')
d2 = np.loadtxt('yspec.out.cow.50')

t = d[:,0]/60
u = d[:,1]
u1 = d1[:,1]
u2 = d2[:,1]

plt.plot(t,u)
plt.plot(t,u-u2)
plt.xlabel('time (min)')
plt.ylabel('velocity (m/s)')
plt.show()
```
